```python
#Part 1-A (plotting the average growth rate for the states )
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

# Read the csv file
df = pd.read_csv("/Users/lifesight/Desktop/GDP_State/ab40c054-5031-4376-b52e-9813e776f65e.csv")

#deleting the row with 2016-17 
df = df[~df['Duration'].str.contains('2016-17')]

#deleting 'items  Description' which has value as 'Growth over previous year'
df = df[~df['Items  Description'].str.contains('GSDP - CURRENT PRICES')]

# Cleaing the data set where column has na values
df = df.dropna(axis=1, how='all')

#Adding the new row for which Average has been calculated
df.loc['Average'] = df[df.columns].mean(axis=0)

#droping first column which has same values  
df = df.drop(['Items  Description'], axis=1)
#Deleting all india GDP
df = df.drop(['All_India GDP'], axis=1)
#Transposing the dataframe
df_transposed = df.T
#removing the rows which has average as NA
df_transposed = df_transposed.dropna(subset=['Average']) # can delete for single bar chart
#sorting the dataframe by average value
df_transposed = df_transposed.sort_values('Average')

#Plotting the bar chart for average GDP growth rate for the indain states during the years
ax = df_transposed.plot(y= 'Average', rot=0,kind='bar', title ="Average GDP of Indain states", figsize=(15, 5), legend=True, fontsize=12)
ax.set_xlabel("State", fontsize=12)
ax.set_ylabel("Average GDP Growth", fontsize=12)
plt.xticks(rotation=90)
plt.show() 
```

```python
#Categorise the states into four groups based on the GDP per capita 
    #(C1, C2, C3, C4, where C1 would have the highest per capita GDP and C4, the lowest)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import glob,os

# Path of the directory where all the csv files exist for all the states
path = '/Users/lifesight/Desktop/GDP_State/GDP_for _States'
all_files = glob.glob(path + "/*.csv")

#taking a empty list
li = []

# Reading file one by one 
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, encoding='unicode_escape')
    
    df = df[['Item','2014-15']]
    df = df.T
    new_header = df.iloc[0]
    df = df[1:]
    df.columns = new_header
    df['State Name'] = filename.split('-')[1]
    li.append(df)

# merging all files into one dataframe
frame = pd.concat(li, axis=0, ignore_index=True ,sort=False)
#Setting index to state name
frame.set_index('State Name',inplace=True)
frame.reset_index(inplace=True)
#sorting the dataframe on the basis of per capita income
frame = frame.sort_values('Per Capita GSDP (Rs.)', ascending=False)
#Getting the quantile value for per capita income
first_Quantile = np.quantile(frame['Per Capita GSDP (Rs.)'], 1)
second_Quantile = np.quantile(frame['Per Capita GSDP (Rs.)'], .85)
third_Quantile = np.quantile(frame['Per Capita GSDP (Rs.)'], .5)
forth_Quantile = np.quantile(frame['Per Capita GSDP (Rs.)'], .2)
#adding new category column 
frame['Category'] = "None"
#dividing states into different category and adding accordingly into the new column
for i in range(0,len(frame.index)):
    if (frame['Per Capita GSDP (Rs.)'].iloc[i] <= first_Quantile) & (frame['Per Capita GSDP (Rs.)'].iloc[i] > second_Quantile):
        frame['Category'].iloc[i] = 'C1'
    elif (frame['Per Capita GSDP (Rs.)'].iloc[i] <= second_Quantile) & (frame['Per Capita GSDP (Rs.)'].iloc[i] > third_Quantile):
        frame['Category'].iloc[i] = 'C2'
    elif (frame['Per Capita GSDP (Rs.)'].iloc[i] <= third_Quantile) & (frame['Per Capita GSDP (Rs.)'].iloc[i] > forth_Quantile):
        frame['Category'].iloc[i] = 'C3'
    elif (frame['Per Capita GSDP (Rs.)'].iloc[i] <= forth_Quantile):
        frame['Category'].iloc[i] = 'C4'

#printing the dataframe with category
newFrame = frame[['State Name','Per Capita GSDP (Rs.)','Category']]
print(newFrame)
print("")

# getting the list of the sub sub sectors which needs to be removed
dropValues = ['Crops', 'Livestock', 'Forestry and logging','Fishing and aquaculture','Primary','Secondary','Trade & repair services',
              'Hotels & restaurants','Railways','Road transport','Water transport','Air transport','Services incidental to transport',
             'Storage','Communication & services related to broadcasting','Tertiary','Taxes on Products','Subsidies on products',
             'Gross State Domestic Product','Population' ,'Per Capita GSDP ']
#drop the rows which has above rows
for i in range(0,len(dropValues)):
    frame = frame.drop(frame.columns[frame.columns.str.contains(dropValues[i])], axis=1)

#Creating 4 different dataframe on the basis of category
dfC1 = frame[frame['Category']=='C1']
dfC2 = frame[frame['Category']=='C2']
dfC3 = frame[frame['Category']=='C3']
dfC4 = frame[frame['Category']=='C4']
```

```python
# Find the top 3/4/5 sub-sectors (such as agriculture, forestry and fishing, crops, manufacturing etc., 
    #not primary, secondary and tertiary) that contribute to approximately 80% of the GSDP of each category.

# plotting and getting top sub-sectors for category C1


#Now it is different dataframe hence removing category column which is extra
dfC1 = dfC1.drop(['Category'], axis=1)
#Setting the index
dfC1.set_index('State Name', inplace=True)
#Creating new row with sum of all the values of categories
dfC1.loc['Sum of Category'] = dfC1[dfC1.columns].sum(axis=0)
#Getting total of all the sub-category's sum of sum
dfC1TotalGSDP = dfC1.loc['Sum of Category'][len(dfC1.columns)-1]
#Adding one more row for the percentage share of sub-category
dfC1.loc['Percentage share of sub-sectors for Category C1'] = ((dfC1.loc['Sum of Category'])/dfC1TotalGSDP)*100
#Setting it to index
dfC1 = dfC1[dfC1.index == 'Percentage share of sub-sectors for Category C1']
#Dropping the extra column
dfC1 = dfC1.drop(['TOTAL GSVA at basic prices'], axis=1)
#Tranposing the resulting dataframe to look it better
dfC1 = dfC1.T
#sorting new data frame on the basis of percentage share of sub-sectors
dfC1 = dfC1.sort_values('Percentage share of sub-sectors for Category C1', ascending=False)
#Calculating cumulative sum for the calculated percentage 
dfC1['Cumulative Sum'] = dfC1['Percentage share of sub-sectors for Category C1'].cumsum(axis = 0) 
#Creating new data frame for category C1 for sub-sectors which contributes approx 80% of GDP
dfC1new = dfC1[dfC1['Cumulative Sum'] < 80] 
# Printing new data frame for category C1 for sub-sectors which contributes approx 80% of GDP
print("Top 3/4/5 sub-sectors that contribute to approximately 80% of the GSDP of C1 category")
print(list(dfC1new.index))
print("")
#Plotting bar chart for the sub-categories shares to the GDP of C1
ax = dfC1.plot(y= 'Percentage share of sub-sectors for Category C1', rot=0,kind='bar', title ="sub-sectors as a percentage of the GSDP of C1", figsize=(25, 10), legend=True, fontsize=12)
ax.set_xlabel("Sub-Sectors", fontsize=15)
ax.set_ylabel("Percentage of the GSDP of C1", fontsize=15)
plt.xticks(rotation=90)
plt.show()
```
